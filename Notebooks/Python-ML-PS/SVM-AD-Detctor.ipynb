```python
import numpy as np
import pandas as pd
```

```python
data_file='ad-dataset/ad.data'
ad_data = pd.read_csv(data_file,sep=",",low_memory=False,header=None)
```

```python
ad_data.head()
```

```python
#function to impute the missing values

def toNum(cell):
    try:
        return np.float(cell)
    except:
        return np.nan
```

```python
#function will apply the toNum function to a col/series of pandas DF

def seriesToNum(series):
    return series.apply(toNum)
```

```python
train_data = ad_data.iloc[0:,0:-1].apply(seriesToNum)
```

```python
train_data.head(11)
```

```python
#dropping rows with  missing values
train_data=train_data.dropna()
train_data.head(11)
```

```python
#function for training labels
def toLabel(name):
    if name == "ad.":
        return 1
    else:
        return 0
```

```python
train_labels = ad_data.iloc[train_data.index,-1].apply(toLabel)
train_labels
```

```python
from sklearn.svm import LinearSVC
```

```python
#fitting a model using 0:2300 cols in the dataset
clf = LinearSVC()
clf.fit(train_data[0:2300],train_labels[0:2300])
```

```python
clf.predict(train_data.iloc[12].reshape(1,-1) ) #predicting ad , return 1
```

```python
clf.predict(train_data.iloc[-1].reshape(1,-1) ) #predicting ad , return 0
```
